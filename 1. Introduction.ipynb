```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from mpl_toolkits.basemap import Basemap
```

Nowadays we cannot imagine life without electricity as it has become one of the most important resources for mankind. We use electrical power every day for countless tasks that we probably do not even realise. Just like every other resource on the planet, the electricity is limited and its price is determined by the ratio of supply and demand. To moderate the price and keep it stable, one needs to provide a regulated amount of electricity. However, estimating what the demand for electricity will be in the near future is not an easy task because many factors influence it. __[Yi-Tui Chen](https://pdfs.semanticscholar.org/171b/b509249400e37df8355dd1b2417b3d6736cd.pdf)__ studied the factors that influence the consumption of electricity in Taiwan's households. It was found that 31% of the yearly usage of electricity is used for air conditioning. AC is also the biggest consumer and second place is cooking with 28%. From that it follows logically that one of the biggest factors that influences electricity consumption is the weather. Therefore, my research question is whether we can predict the power consumption by using only data about the weather.

To simplify the task, I focused on 6 major provinces in the state of India. That way I am effectively repeating the same research 6 times on independent regions which will give confidence to the findings. The cities of interest are shown below.

```python
cities = pd.read_csv("data/cleaned/CitiesLocation.csv")
cities
```

```python
plt.subplots(figsize = (20, 15))
map = Basemap(width = 1200000, height = 900000, projection = 'lcc', resolution= 'l',
                    llcrnrlon = 67, llcrnrlat = 5, urcrnrlon = 99, urcrnrlat = 37, lat_0 = 28, lon_0 = 77)
map.drawmapboundary()
map.drawcountries()
map.drawcoastlines()
lg = np.array(cities['longitude'])
lt = np.array(cities['Latitude'])
nc = np.array(cities['Location_Name'])
x, y = map(lg, lt)
plt.scatter(x, y, s = 75, marker = "*")
for ncs, xpt, ypt in zip(nc, x, y):
    plt.text(xpt + 60000, ypt + 30000, ncs, fontsize = 13, fontweight = 'bold')
plt.title('Cities of interest',fontsize = 20)
plt.show()
```
